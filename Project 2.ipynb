```python
import numpy as np
import copy
import math
import heapq
import time
import matplotlib.pyplot as plt
import cv2

#Inputs
x_start=int(input("Initial x coordinate="))
y_start=int(input("Initial y coordinate="))
x_goal=int(input("Goal x coordinate="))
y_goal=int(input("Goal y coordinate="))
start=(x_start,y_start) 
goal=(x_goal,y_goal)   
Maximum_size_x = 400
Maximum_size_y = 250

#Moving Up 
def Up(curr_node):
    x = curr_node[0]
    y = curr_node[1]
    curr_node = (x,y)
    new_node=()
    new_node_y = curr_node[0]
    new_node_y-=1
    new_node = (new_node_y,curr_node[1])
    if new_node[0]>=0 and new_node[1]>=0:
        return(new_node,True)
    else:
        return(curr_node,False)
#Down      
def Down(curr_node):
    x = curr_node[0]
    y = curr_node[1]
    curr_node = (x,y)
    new_node=()
    new_node_y = curr_node[0]
    new_node_y+=1
    new_node = (new_node_y,curr_node[1])
    if new_node[0]>=0 and new_node[1]>=0:
        return(new_node,True)
    else:
        return(curr_node,False)
#Left      
def Left(curr_node):
    x = curr_node[0]
    y = curr_node[1]
    curr_node = (x,y)
    new_node=()
    new_node_x = curr_node[1]
    new_node_x-=1
    new_node = (curr_node[0],new_node_x)
    if new_node[0]>=0 and new_node[1]>=0:
        return(new_node,True)
    else:
        return(curr_node,False)
#Right   
def Right(curr_node):
    x = curr_node[0]
    y = curr_node[1]
    curr_node = (x,y)
    new_node=()
    new_node_x = curr_node[1]
    new_node_x+=1
    new_node = (curr_node[0],new_node_x)
    if new_node[0]>=0 and new_node[1]>=0:
        return(new_node,True)
    else:
        return(curr_node,False)

#Top left  
def TLeft(curr_node):
    x = curr_node[0]
    y = curr_node[1]
    curr_node = (x,y)
    new_node=()
    new_node_y = curr_node[0]
    new_node_y-=1
    new_node_x = curr_node[1]
    new_node_x-=1
    new_node = (new_node_y,new_node_x)
    if new_node[0]>=0 and new_node[1]>=0:
        return(new_node,True)
    else:
        return(curr_node,False)
#Top right
def TRight(curr_node):
    x = curr_node[0]
    y = curr_node[1]
    curr_node = (x,y)
    new_node=()
    new_node_y = curr_node[0]
    new_node_y-=1
    new_node_x = curr_node[1]
    new_node_x+=1
    new_node = (new_node_y,new_node_x)
    if new_node[0]>=0 and new_node[1]>=0:
        return(new_node,True)
    else:
        return(curr_node,False)
#Bottom left
def BLeft(curr_node):
    x = curr_node[0]
    y = curr_node[1]
    curr_node = (x,y)
    new_node=()
    new_node_y = curr_node[0]
    new_node_y+=1
    new_node_x = curr_node[1]
    new_node_x-=1
    new_node = (new_node_y,new_node_x)
    if new_node[0]>=0 and new_node[1]>=0:
        return(new_node,True)
    else:
        return(curr_node,False)
#Bottom right
def BRight(curr_node):
    x = curr_node[0]
    y = curr_node[1]
    curr_node = (x,y)
    new_node=()
    new_node_y = curr_node[0]
    new_node_y+=1
    new_node_x = curr_node[1]
    new_node_x+=1
    new_node = (new_node_y,new_node_x)
    if new_node[0]>=0 and new_node[1]>=0:
        return(new_node,True)
    else:
        return(curr_node,False)

# List of all coordinate points on the obstacle space
all_points = []
for i in range(0,401): 
    for j in range(251): 
        all_points.append((i,j)) 
        
#Generating graph
def generateGraph(start,size_x,size_y): 
    i = start[0] 
    j = start[1] 
    if i <size_x and j<size_y:
        graph={}
        
        if i==0 and j==0: 
            graph[(i,j)]={(i+1,j+1),(i+1,j),(i,j+1)}
        
        elif i==size_x-1 and j==size_y-1:
            graph[(i,j)]={(i-1,j),(i-1,j-1),(i,j-1)}
        
        elif i==size_x-1 and j ==0:
            graph[(i,j)]={(i-1,j),(i-1,j+1),(i,j+1)}
        
        elif j==size_y-1 and i ==0:
            graph[(i,j)]={(i,j-1),(i+1,j-1),(i+1,j)}
        #For points along the borders
        elif i == 0 and j!=0 and j!=size_y-1:
            graph[(i,j)]={(i,j-1),(i,j+1),(i+1,j-1),(i+1,j),(i+1,j+1)}
       
        elif i == size_x-1 and j!=0 and j!=size_y-1:
            graph[(i,j)]={(i,j-1),(i,j+1),(i-1,j-1),(i-1,j),(i-1,j+1)}
        
        elif j == 0 and i!=0 and i!=size_x-1:
            graph[(i,j)]={(i-1,j),(i+1,j),(i+1,j+1),(i,j+1),(i-1,j+1)}
        
        elif j == size_y-1 and i!=0 and i!=size_x-1:
            graph[(i,j)]={(i-1,j),(i+1,j),(i+1,j-1),(i,j-1),(i-1,j-1)} 
        
        else: 
            graph[(i,j)]={(i-1,j),(i-1,j+1),(i-1,j-1),(i+1,j-1),(i+1,j),(i+1,j+1),(i,j-1),(i,j+1)}
        
        return(graph)
    else:
        pass
    
def costCalculationatAllNodes(graph,start):
    new_dic={}
    for key,value in graph.items():
        new_dic[key]={}
        for neighbour in value:
            U =Up(key)
            D =Down(key)
            L =Left(key)
            R =Right(key)
            TL = TLeft(key)
            TR =TRight(key)
            BL = BLeft(key)
            BR = BRight(key)
            if (neighbour==R[0]) or (neighbour==L[0]) or (neighbour==U[0]) or (neighbour==D[0]):
                new_dic[key][neighbour]=1          
            elif (neighbour==TL[0]) or (neighbour==TR[0]) or (neighbour==BL[0]) or (neighbour==BR[0]):
                new_dic[key][neighbour]=1.414 
    return(new_dic)

all_distance = {}
backtracking = {}
visited = []
check=0
def dijkstraAlgorithm(graph,start):
    global check
    global visited
    all_distance[start]=0
    visited.append(start)
    for vertex,edge in graph.items():
        all_distance[vertex]=math.inf
    priority_queue = [(0,start)]
    while len(priority_queue)>0 and check!=[]:
        curr_dist,curr_vert = heapq.heappop(priority_queue) 
        if curr_dist>all_distance[curr_vert]:
            continue
        for neighbour,cost in graph[curr_vert].items():
            distance = curr_dist + cost 
            if distance < all_distance[neighbour]:
                backtracking[neighbour]={}
                backtracking[neighbour][distance]=curr_vert
                all_distance[neighbour]=distance
                heapq.heappush(priority_queue, (distance, neighbour))
                if neighbour not in visited:
                    visited.append(neighbour)
                    if neighbour==goal:
                        print('Found Goal')
                        check=[]
                        break
    return(all_distance,visited,backtracking)     

# Backtracking
def BackTrack(backtrack_dict,goal,start):
    back_track_list = []
    back_track_list.append(start)
    while goal!=[]:
        for k,v in backtracking.items():
            for k2,v2 in v.items():
                if k==start:
                    if v2 not in back_track_list:
                        back_track_list.append(start)
                    start=v2
                    if v2==goal:
                        goal=[]
                        break
    return(back_track_list)

#Finding and plotting obstacles 
def PointRobotdijkstra(Maximum_size_x,Maximum_size_y,start,goal):
    Maximum_size_x+=1 
    Maximum_size_y+=1
    all_points = []
    for i in range(400): 
        for j in range(250): 
            all_points.append((i,j)) 
    print('Length of all points')
    print(len(all_points))
    list_of_all_points = []
    r = 40
    for c in all_points:
        x = c[0]
        y = c[1]
        #circle shaped obstacle
        if(((y -185)**2 + (x-300)**2 - (r+5)**2)<=0):
            list_of_all_points.append((x,y))
        #polygon shaped obstacle
        s1 = (y-5) - ((0.316) *(x+5)) - 173.608  
        s2 = (y+5) + (1.23 * (x+5)) - 229.34 
        s3 = (y-5) + (3.2 * (x-5)) - 436 
        s4 = (y+5) - 0.857*(x-5) - 111.42 
        s5 = y + (0.1136*x) - 189.09
        
        #hexagon shaped obstacle
        h1 = (y-5) - 0.577*(x+5) - 24.97
        h2 = (y-5) + 0.577*(x-5) - 255.82
        h3 = (x-6.5) - 235 
        h6 = (x+6.5) - 165 
        h5 = (y+5) + 0.577*(x+5) - 175 
        h4 = (y+5) - 0.577*(x-5) + 55.82
        if(h1<0 and h2<0 and h3<0 and h4>0 and h5>0 and h6>0) or (s1<0 and s5>0 and s4>0)or (s2>0 and s5<0 and s3<0):
            list_of_all_points.append((x,y))
       
    if goal in list_of_all_points:
        print('Goal is in obstacle, try again!')
    
    print(' Length of all points with the shape is : ')
    print(len(list_of_all_points))
    base_graph = {}
    for i in range(Maximum_size_x-1,-1,-1):
        for j in range(Maximum_size_y-1,-1,-1):
            graph = generateGraph((i,j),Maximum_size_x,Maximum_size_y)
            base_graph[(i,j)]=graph[(i,j)]
    print('Length of graph before removing:')
    print(len(base_graph))
    for key,value in base_graph.items():
        value_copy = value.copy()
        for coordinates in value_copy:
            if coordinates in list_of_all_points:
                value.remove(coordinates) 
    base_graph_copy=base_graph.copy()
    for key,value in base_graph_copy.items():
        if key in list_of_all_points:
            del base_graph[key]
    print('Length of graph after removing:')
    print(len(base_graph))
    costs_calculated = costCalculationatAllNodes(base_graph,start)
    actual_graph = costs_calculated
    all_distance = {}
    backtracking = {}
    visited = []
    all_distance,visited,backtracking= dijkstraAlgorithm(actual_graph,start) #can alter the start here
    all_distance_copy = all_distance.copy()
    for k,v in all_distance_copy.items():
        if all_distance_copy[k] == math.inf:
            del all_distance[k]
    return(all_distance,visited,backtracking,list_of_all_points)


all_distance,visited,backtrack,listofallpointsformap= PointRobotdijkstra(Maximum_size_x,Maximum_size_y,start,goal)

len(backtrack)
backtracked_final = BackTrack(backtrack,start,goal)
print(backtracked_final)

#Displaying map
new_canvas = np.zeros((251,401,3),np.uint8) 

for c in listofallpointsformap: 
    x = c[1]
    y = c[0]
    new_canvas[(x,y)]=[0,255,255] 

new_canvas = np.flipud(new_canvas)
new_canvas_copy_backtrack = new_canvas.copy()
new_canvas_copy_visited = new_canvas.copy()
new_canvas_copy_visited = cv2.resize(new_canvas_copy_visited,(600,400))
cv2.imshow('Map',new_canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()

#Animation
pygame.init()

display_width = 400
display_height = 250

gameDisplay = pygame.display.set_mode((display_width,display_height),pygame.FULLSCREEN)
pygame.display.set_caption('Dijkstra Animation')

black = (0,0,0)
white = (0,255,255)
surf = pygame.surfarray.make_surface(new_canvas_copy_visited)

clock = pygame.time.Clock()
done = False
while not done:
    for event in pygame.event.get():   
        if event.type == pygame.QUIT:  
            done = True   
 
    gameDisplay.fill(black)
    for path in visited:
        if path not in new_canvas_copy_visited:
            x = path[0]
            y = abs(250-path[1])
            pygame.draw.rect(gameDisplay, white, [x,y,1,1])
            pygame.display.flip()
    for path in backtracked_final:
        
        pygame.time.wait(5)
        x = path[0]
        y = abs(250-path[1])
        pygame.draw.rect(gameDisplay, (0,0,255), [x,y,1,1])
        pygame.display.flip()
        
    done = True
pygame.quit()


for path in visited:
    x = path[0]
    y = path[1]
    new_canvas_copy_backtrack[(250-y,x)]=[255,0,0] 
new_backtracked = cv2.resize(new_canvas_copy_backtrack,(800,500))
cv2.imshow('Solved',new_backtracked)
cv2.waitKey(0)
cv2.destroyAllWindows()

for path in backtracked_final:
    x = path[0]
    y = path[1]
    new_canvas_copy_backtrack[(250-y,x)]=[0,255,0] 
new_backtracked = cv2.resize(new_canvas_copy_backtrack,(800,500))
cv2.imshow('backtracked',new_backtracked)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
